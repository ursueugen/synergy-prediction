```python
import pandas as pd
import networkx as nx
```

```python
df = pd.read_csv("BIOGRID-ORGANISM-Caenorhabditis_elegans-4.4.203.tab3.txt", sep='\t')
print(df.columns)
df.head(3)
```

```python
GENE_NAME_A_COL = "Official Symbol Interactor A"
GENE_NAME_B_COL = "Official Symbol Interactor B"

genes_biogrid = set(df[GENE_NAME_A_COL]).union(df[GENE_NAME_B_COL])
len(genes_biogrid)
```

```python
df_syn = pd.read_csv("synergyage_database_augmented.tsv", sep='\t')
genes_syn = set([])
for genes_str in df_syn["genes_str"]:
    genes_syn = genes_syn.union( set( genes_str.split(";") ) )

raw_intersection_perc = len(genes_syn.intersection(genes_biogrid)) / len(genes_syn)
print(f"Only {raw_intersection_perc} of synergyage genes are recalled in biogrid on a naive merge.")
```
